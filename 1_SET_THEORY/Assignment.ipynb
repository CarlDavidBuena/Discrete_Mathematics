```python
# Problem 1: Symmetric Difference of C and D
print("Problem 1: Find C ⊕ D (Symmetric Difference of C and D)")
print("\nSolution 1:")
print("The symmetric difference of two sets C and D, denoted as C ⊕ D, is the set of elements that are in either C or D, but not in both.")
print("C ⊕ D = (C ∪ D) - (C ∩ D)")

# Calculate it:
print("\nNow, let's calculate it:")
print("C ∪ D =", C.union(D))
print("C ∩ D =", C.intersection(D))
symmetric_difference_CD = C.symmetric_difference(D)
print("C ⊕ D =", symmetric_difference_CD)

# Problem 2: Set difference of A and B
print("\nProblem 2: Find A \ B (Set difference of A and B)")
print("\nSolution 2:")
print("The set difference of two sets A and B, denoted as A \ B, is the set of elements that are in A but not in B.")
print("A \ B =", A.difference(B))

# Problem 3: Set difference of B and A
print("\nProblem 3: Find B \ A (Set difference of B and A)")
print("\nSolution 3:")
print("The set difference of two sets B and A, denoted as B \ A, is the set of elements that are in B but not in A.")
print("B \ A =", B.difference(A))

# Problem 4: Symmetric Difference of E and F
print("\nProblem 4: Find E ⊕ F (Symmetric Difference of E and F)")
print("\nSolution 4:")
print("Using the same formula as in Problem 1:")
print("E ⊕ F = (E ∪ F) - (E ∩ F)")
print("E ⊕ F =", E.symmetric_difference(F))
```
